```python
# Typecasting 
df['DATE-REPORTED'] = pd.to_datetime(df['DATE-REPORTED'],format=r'%d-%m-%Y')
df['DISBURSED-DT'] = pd.to_datetime(df['DISBURSED-DT'],format=r'%d-%m-%Y')
df['CLOSE-DT'] = pd.to_datetime(df['CLOSE-DT'],format=r'%d-%m-%Y',errors = 'ignore')
df['LAST-PAYMENT-DATE'] = pd.to_datetime(df['LAST-PAYMENT-DATE'],format=r'%d-%m-%Y')
```

```python
#Finding Customer vintage
df_customer_vintage = df[['LOS-APP-ID','DATE-REPORTED']]
df_customer_vintage = df_customer_vintage.groupby(by=['LOS-APP-ID']).agg({'DATE-REPORTED':['min']})
df_customer_vintage.reset_index(inplace = True)
df_customer_vintage[df_customer_vintage['LOS-APP-ID']==4]
df_customer_vintage['Vintage'] = df_customer_vintage.apply(lambda x : today- x['DATE-REPORTED'], axis=1)
df_customer_vintage.to_csv('Customer_vintage.csv', index=False)
df_customer_vintage.head()
```

```python
#Assign Loan Type

def assign_loan_type (var1):
    secured = ['Gold Loan','Business Loan - Secured','Business Loan Against Bank Deposits','GECL Loan secured','Gold Loan','Housing Loan','Loan Against Shares/Securities',
               'Loan Against Bank Deposits','Property Loan','Microfinance - Housing Loan','JLG Group','JLG Individual','Mudra Loans - Shishu / Kishor / Tarun','Pradhan Mantri Awas Yojana - CLSS','Property Loan']
    if var1 in secured:
        return 'Secured'
    else:
        return 'Unsecured'
    
df['Secured/Unsecured'] = df['ACCT-TYPE'].apply(assign_loan_type)
```

```python
# Excluding Report date more than 3 yrs ago

df['To be included'] = df.apply(lambda x : 1 if x['DATE-REPORTED'].year > 2020 else 0 , axis=1)
```

```python
#Account status

def current_status_func (var1):
    default = ['Sold/Purchased','WRITTEN OFF','Written Off','WILFUL DEFAULT','SUIT FILED (WILFUL DEFAULT)','Delinquent','SUIT FILED','Suit Filed']
    if var1 in default:
        return 'Default'
    elif var1 in ['CLOSED','SETTLED','Closed','Settled','Post Write Off Closed','Post Write Off Settled','Restructured  and  Closed','Cancelled','Loan Declined']:
        return 'Closed'
    elif var1 in ['ACTIVE','Active','RESTRUCTURED','Restructured']:
        return 'Active'
    else:
        print('Other values of account type detected. Add in the category above', var1)
        return var1

df['Current_Status'] = df['ACCOUNT-STATUS'].apply(current_status_func)



#Self indicator
df['Own']= df['SELF-INDICATOR'].apply(lambda x : 1 if x ==True else 0)
print('Own tradelines', df['Own'].sum())
df.head()
```

```python
#Dropping where only tradeline is ignore tradeline
df_final['Used tradelines']= df_final['Secured_Closed_Own']+df_final['Secured_Active_Own']+df_final['Secured_Default_Own']+df_final['Unsecured_Closed_Own']+df_final['Unsecured_Active_Own']+df_final['Unsecured_Default_Own']+df_final['Secured_Closed_Rest']+df_final['Secured_Active_Rest']+df_final['Secured_Default_Rest']+df_final['Unsecured_Closed_Rest']+df_final['Unsecured_Active_Rest']+df_final['Unsecured_Default_Rest'] 
df_final2 = df_final[df_final ['Used tradelines']>0]
```

```python
# Assigning default category

def cetegory(df):
    if df['Secured_Active_Rest']+df['Secured_Default_Rest']+df['Unsecured_Active_Rest']+df['Unsecured_Default_Rest']==0:
        return 'Only Seller'
    elif df['Secured_Active_Rest']+df['Secured_Default_Rest']==0:
        return 'Only Unsecured'
    elif df['Unsecured_Active_Rest']+df['Unsecured_Default_Rest']==0:
        return 'Only Secured'
    else:
        return 'Both'

df_final2['Category'] = df_final2.apply(cetegory, axis=1)

# Calculating percentage of Other secured default

def other_secured_default(df):
    if df['Secured_Active_Rest']+df['Secured_Default_Rest'] ==0:
        return 0
    else:
        val = df['Secured_Default_Rest']/(df['Secured_Active_Rest']+df['Secured_Default_Rest'])
        return val

df_final2['Other secured Default'] = df_final2.apply(other_secured_default, axis=1)


# Calculating percentage of Other unsecured default

def other_unsecured_default(df):
    if df['Unsecured_Active_Rest']+df['Unsecured_Default_Rest'] ==0:
        return 0
    else:
        val = df['Unsecured_Default_Rest']/(df['Unsecured_Active_Rest']+df['Unsecured_Default_Rest'])
        return val

df_final2['Other unsecured Default'] = df_final2.apply(other_unsecured_default, axis=1)



# Calculating percentage of Other secured default percentage

def other_secured_default_percentage(df):
    
    if df['Secured_Active_Rest']+df['Secured_Default_Rest'] ==0:
        return ""
    
    elif (df['Secured_Default_Rest']/(df['Secured_Active_Rest']+df['Secured_Default_Rest']))==0:        
        return 'No default'
    
    elif (df['Secured_Default_Rest']/(df['Secured_Active_Rest']+df['Secured_Default_Rest']))<=0.5:        
        return 'Default less than 50%'
    else:
        return 'Default more than 50%'


df_final2['Other secured Default_bin'] = df_final2.apply(other_secured_default_percentage, axis=1)


# Calculating percentage of Other unsecured default percentage

def other_unsecured_default_percentage(df):
    
    if df['Unsecured_Active_Rest']+df['Unsecured_Default_Rest'] ==0:
        return ""
    elif (df['Unsecured_Default_Rest']/(df['Unsecured_Active_Rest']+df['Unsecured_Default_Rest']))==0:        
        return 'No default'
    elif (df['Unsecured_Default_Rest']/(df['Unsecured_Active_Rest']+df['Unsecured_Default_Rest']))<=0.5:        
        return 'Default less than 50%'
    else:
        return 'Default more than 50%' 

df_final2['Other unsecured Default_bin'] = df_final2.apply(other_unsecured_default_percentage, axis=1)

# Final Bucket

def final_tagging(df):
    if df['Category']=='Only Seller':
        return 'Only Seller'
    elif df['Category']=='Only Unsecured':
        return 'Only Unsecured_'+str(df['Other unsecured Default_bin'])
    elif df['Category']=='Only Secured':
        return 'Only Unsecured_'+str(df['Other secured Default_bin'])
    else:
        return "Secured_"+str(df['Other secured Default_bin'])+"_Unsecured_"+str(df['Other unsecured Default_bin'])




df_final2['Final tag'] = df_final2.apply(final_tagging, axis=1)
df_final2.head(2)
```
